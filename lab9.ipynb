# **Lab 9 : Boucle For**

## **Lab9. exercice 1 : Parcours de base**


```python
# 1. Affichez les nombres de 0 à 9 en utilisant range()
print("les nombres de 0 à 9")
for i in range(10):
    print(i)

# 2. Affichez les nombres pairs entre 0 et 20
print("\n les nombres de 0 à 20 par 2")
for i in range(0, 21, 2):
    print(i)

# 3. Parcourez la liste suivante et affichez chaque élément
print("\n les fruits")
fruits = ["pomme", "banane", "orange", "kiwi"]
for fruit in fruits:
    print(fruit)
```

## **Lab9. exercice 2 : Manipulation de chaînes**


```python
# 1. Comptez le nombre de 'a' dans le mot "abracadabra"
mot = "abracadabra"
count = 0

for lettre in mot:
    if lettre == 'a':
        count += 1

print(f"Nombre de 'a': {count}")

# 2. Inversez une chaîne de caractères (sans utiliser [::-1])
chaine = "Python"
inverse = ""

for c in chaine:
    inverse = c + inverse

print(inverse)
```

## **Lab9. exercice 3 : Listes et for**


```python
# 1. Créez une liste des carrés des nombres de 1 à 10
carres = []
for i in range(1, 11):
    carres.append(i**2)

print(carres)

# 2. Filtrez les nombres pairs d'une liste
nombres = [12, 15, 8, 9, 24, 3, 18]
pairs = []
for n in nombres:
    if n % 2 == 0:
        pairs.append(n)

print(pairs)
```

## **Lab9. exercice 4 : break et continue**


```python
# 1. Trouvez le premier nombre divisible par 7 entre 100 et 200
for n in range(100, 201):
    if n % 7 == 0:
        print(f"Premier divisible par 7: {n}")
        break

# 2. Affichez les nombres de 1 à 10 sauf 5 et 7
for i in range(1, 11):
    if i == 5 or i == 7:
        continue
    print(i)
```
